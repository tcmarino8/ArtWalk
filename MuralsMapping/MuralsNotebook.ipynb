```python
import matplotlib.pyplot as plt
import numpy as np
import pykml
from pykml import parser

import geopandas as gpd
from shapely.geometry import Point, Polygon
import pandas as pd

import folium
```

```python
kml_file = "Murals of Berkeley.kml"
with open(kml_file) as f:
    doc = parser.parse(f)
```

```python
gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'
my_murals = gpd.read_file("Murals of Berkeley.kml", driver='KML')
murals_points = [[point.xy[1][0], point.xy[0][0]] for point in my_murals.geometry]
my_murals
```

```python
map = folium.Map(location=[murals_points[0][0], murals_points[0][1]], tiles="OpenStreetMap", zoom_start=12)

for i, coodinates in enumerate(murals_points):
    map.add_child(folium.Marker(
        location=coodinates,
        popup = 'Mural: '+
        str(my_murals['Name'][i])
        )    
    )
map
```
